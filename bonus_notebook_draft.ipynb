```python
import pandas as pd
from plotnine import ggplot, aes, geom_line, labs

# Make sure your CSV file is in the correct path
file_path = 'data/Figure5_2.csv'
energy_data = pd.read_csv(file_path)

# Melting the DataFrame from wide format to long format
energy_data_long = pd.melt(energy_data, id_vars='Year', var_name='Energy Source', value_name='Consumption')

energy_data_long.head()

# Plotting with plotnine
plot = (ggplot(energy_data_long, aes(x='Year', y='Consumption', color='Energy Source'))
        + geom_line()
        + labs(title='Energy Consumption by Source Over Time',
               x='Year',
               y='Energy Consumption')
        )

# Display the plot
print(plot)
```
